# Main Process

```python
def FPMiner(profiles_data, relationships_data, minsup, directed_=True):
  MODIFY_INDEX = 0
  if directed_ == True:
    g = directed.MyGraph()
  else:
    g = undirected.MyGraph()
  print('Data Loading ...')
  g.loadData(profiles_data, relationships_data)
  print('Mining Frequent patterns ....')
  fp = g.FPMiner(minsup)
  return fp

def ruleGen(fp, conf=0.9):
  print('Generating association rules ...')
  rules, confs = fp.ruleGen(conf)
  return rules, confs
```

# Directed

## example

```python
g = directed.MyGraph()
profiles = 'DATASETS/Example/example_profiles.txt'
relationships = 'DATASETS/Example/example_relationships.txt'
g.loadData(profiles, relationships)
g.plot()
```

```python
# s = [2, 3]
start = time.time()
fc = FPMiner(profiles, relationships, 3, True)
print('Runtime: ', time.time()-start)
print(fc.getNum())
fc.display()
```

```python
start = time.time()
rules, confs = ruleGen(fc, 0.9)
print('Runtime: ', time.time()-start)
print(len(rules))#, rules, confs)
```
